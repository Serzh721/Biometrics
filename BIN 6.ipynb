##  I. Image Recognition

```python
from imageai.Prediction import ImagePrediction
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

execution_path = os.getcwd()
prediction = ImagePrediction()
prediction.setModelTypeAsResNet()
prediction.setModelPath( execution_path + "/resnet50_weights_tf_dim_ordering_tf_kernels.h5")
prediction.loadModel()
```

```python
images = ["/img/plane.jpg", "/img/car.jpg", "/img/lion.jpg", "/img/bee.jpg", "/img/apple.jpg", "/img/basketball.jpg", "/img/Dobby.jpg"]
names = ['Самолет', 'Машина', 'Лев', 'Пчела', 'Яблоко', 'Баскетбольный мяч', 'Добби']
for img, name in zip(images, names):
    print(name)
    path_to_img = execution_path + img
    img_to_draw = mpimg.imread(path_to_img)
    imgplot = plt.imshow(img_to_draw)
    plt.show()
    predictions, percentage_probabilities = prediction.predictImage(path_to_img, result_count=5)
    for index in range(len(predictions)):
        print(predictions[index] , " : " , percentage_probabilities[index])
    print()
```

#### Выводы:
- Самолет, льва, пчелу и баскетбольный мяч библиотека идентифицировала безупречно, с большим перевесом в их пользу.
- К моему удивлению, машина распозналась плохо (на первом месте оказалась решетка радиатора). Хотя дальше можно увидеть разные виды машин, даже кабриолет, но в точку попасть не удалось. Думаю, что это из-за довольно специфичного вида машины.
- С яблоками вышла интересная ситуация. Библиотека не смогла точно распознать что изображено. Предположив даже, что это банан, клубника, лимон или ожерелье. Совсем не в ту степь. ОДНАКО! На первом месте стоит Granny Smith. Прогуглив, оказалось, что это яблоки, но этот сорт зеленый. Уж ладно, простим:)
- Распознавание Добби было исключительно в рамках любопытства. Угадать, конечно, не получилось. Предположения здесь очень интересные. В целом, точно есть такие маски. Чтобы понять, что находится на втором месте, опять пришлось спросить гугл. Оказалось, что это пород собаки и он действительно смахивает на Добби. Удивительно!

## II. Object Detection

```python
from imageai.Detection import ObjectDetection
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

execution_path = os.getcwd()
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath( execution_path + "/resnet50_coco_best_v2.0.1.h5")
detector.loadModel()
```

```python
images = ["\img\plane.jpg", "/img/car.jpg", "/img/lion.jpg", "/img/bee.jpg", "/img/apple.jpg", "/img/basketball.jpg", "/img/Dobby.jpg", "/img/metro.jpg", "/img/people1.jpg", "/img/people2.jpg", "/img/traffic.jpg", "/img/police.jpeg"]
names = ['Самолет', 'Машина', 'Лев', 'Пчела', 'Яблоко', 'Баскетбольный мяч', 'Добби', 'Метро', 'Люди на улице', 'Свадьба', 'Пробка', 'Полиция']
new_images = ["\out\plane.jpg", "/out/car.jpg", "/out/lion.jpg", "/out/bee.jpg", "/out/apple.jpg", "/out/basketball.jpg", "/out/Dobby.jpg", "/out/metro.jpg", "/out/people1.jpg", "/out/people2.jpg", "/out/traffic.jpg", "/out/police.jpeg"]

def draw(path_to_img):
    img_to_draw = mpimg.imread(path_to_img)
    imgplot = plt.imshow(img_to_draw)
    plt.show()
    
def detect_and_draw(name, to_draw_input, to_draw_output):
    print(name)
    detections = detector.detectObjectsFromImage(input_image=to_draw_input, output_image_path=to_draw_output)
    draw(to_draw_output)
    for eachObject in detections:
        print(eachObject["name"] , " : " , eachObject["percentage_probability"] )
    print()

for img, new_img, name  in zip(images, new_images, names):
    to_draw_input = execution_path + img
    to_draw_output = execution_path + new_img
    detect_and_draw(name, to_draw_input, to_draw_output)
```

#### Выводы:
- Несовместимость с последними версиями библиотеки Tensorflow
- Стоит отметить, что объекты обнаружены прекрасно, но есть небольшие проблемы с определением найденного
- С самолетом вновь не возникло проблем. Также там нашелся некий пожарный гидрант, хотя это кусочек фонаря 
- Машину распознал как грузовичок. Ну пусть будет так.
- Лев почему-то стал собакой...
- Пчелу не идентифицировал, а нашел кошку, и птицу
- Яблоки распознал, но нашел 1 лишнее
- С мячем все ок
- Добби распознал как человека. На плече действительно узелок похож на перчатку :) 

Добавил 5 картинок с множеством объектов (на конференциях в Zoom подчеркивалось это требование)

- довольно хорошо распознал людей (кроме фото со свадьбы) и машины в пробке
- работает общее "правило" чем ближе объект - тем лучше распознает(см. числовой результат), НО требуется достаточная видимость (касательно моих примеров понятно - чем дальше объект, тем он меньше и тем сильнее его закрывают объекты, находящиеся на переднем плане). Проследить это можно на фотографии с полицией - объекты крупные, четкие - результат отличный. На свадебной фотографии все стоят очень плотно, у некоторых людей видно только головы - не распознал, хотя у мужчин в первых рядах распознаны даже галстуки!

## III. Video Detection and Analysis

```python
from imageai.Detection import VideoObjectDetection
import os

execution_path = os.getcwd()
detector = VideoObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath(execution_path +"/yolo.h5")
detector.loadModel()

video_path = detector.detectObjectsFromVideo(input_file_path=execution_path+"/traffic-mini.mp4", 
                                             output_file_path=execution_path + "/out/traffic_mini_detected_1", 
                                             frames_per_second=29, log_progress=True)
```

#### Выводы:
